```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns #gabole
%matplotlib inline

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn import svm
from sklearn.metrics import accuracy_score
from sklearn import preprocessing
from sklearn.metrics import classification_report
from sklearn.model_selection import GridSearchCV
import scipy.optimize as opt

import matplotlib.image as mpimg
from sklearn import tree

#Untuk memberi tampilan yang menarik
class color:
   PURPLE = '\033[95m'
   CYAN = '\033[96m'
   DARKCYAN = '\033[36m'
   BLUE = '\033[94m'
   GREEN = '\033[92m'
   YELLOW = '\033[93m'
   RED = '\033[91m'
   BOLD = '\033[1m'
   UNDERLINE = '\033[4m'
   GREYBACKBLACK =  '\033[0;30;47m'
   END = '\033[0m'

#Memanggil file csv heart dan keterangan dari tiap variabel pada csv heart
df = pd.read_csv('/content/heart.csv')
print(color.BOLD + color.GREYBACKBLACK + "Berikut adalah dataset yang akan kita gunakan"+color.END)
display(df.head())
print("")
print(color.BOLD + color.DARKCYAN + "Keterangan tiap variabel"+color.END)
#with open('/content/about dataset.txt', 'r') as aboutdf:
  #print(aboutdf.read())
print(color.BOLD+"Periksa isi data apakah sudah siap digunakan"+color.END)
print("Terdapat data hilang? ",df.isnull().values.any())
print("Apakah terdapat outliers?")
Q1 = np.quantile(df, .25)
Q3 = np.quantile(df, .75)
IQR = Q3 - Q1
min_IQR = Q1 - 1.5 * IQR
max_IQR = Q3 + 1.5 * IQR
nilai_min = np.min(df)
nilai_max = np.max(df)
if (nilai_min.any() < min_IQR):
    print('Ditemukan low outlier!')
else:
    print('Tidak ditemukan low outlier!')  
if (nilai_max.any() > max_IQR):
    print('Ditemukan high outlier!')
else:
    print('Tidak ditemukan high outlier!')
print("Data sudah siap digunakan dengan variabel targetnya adalah kolom "+color.BOLD+"output"+color.END)
print("")

#Feature Selection with Pearson Correlation
print(color.BOLD + color.BLUE + "Lakukan Seleksi Fitur dengan Pearson Correlation"+color.END)
plt.figure(figsize=(12,12))
korelasi = df.corr()
display(korelasi)
sns.heatmap(korelasi, annot=True, cmap=plt.cm.Blues)
plt.show()

#Akan dicari variabel-variabel yang berkorelasi dengan variabel target (>rata-rata korelasi)
print(color.BOLD + color.RED + "Didapat nilai absolut korelasi antar variabel dengan variabel output adalah sebagai berikut"+color.END)
print(abs(korelasi["output"]))
korelasi["output"]=korelasi["output"][:-1]
bataskor=abs(korelasi["output"][:-1]).mean()
print("")
print("Mean dari korelasi tiap variabel yang ada adalah "+color.BOLD+ "{0}".format(bataskor)+color.END)
print("Sehingga,"+color.BOLD+color.DARKCYAN+ " variabel yang dipilih sebagai hasil seleksi fitur adalah"+color.END)
hasilseleksi = abs(korelasi["output"])[abs(korelasi["output"])>bataskor]
print(hasilseleksi)

#mengolah data awal hingga split dengan perbandingan 2:8
data_x = df[['cp', 'thalachh', 'exng', 'oldpeak', 'slp', 'caa', 'thall']].values
data_y = df['output']
data_xp = preprocessing.StandardScaler().fit(data_x).transform(data_x) #normalisasi data
X_train, X_test, Y_train, Y_test = train_test_split(data_xp, data_y, test_size=0.2, random_state = 4)
print("")

#Metode Logistic Regression
print("--------------------------------------------------------------")
print(color.BOLD + color.RED + "Metode Logistic Regression"+color.END)
grid={"C":np.arange(0.1,1,0.1)}
LR_cv=GridSearchCV(LogisticRegression(),grid,cv=2,scoring='accuracy')
LR_cv.fit(X_train, Y_train)
print("Akan dipilih parameter terbaik untuk model")
print("Parameter (C) Terbaik: ",LR_cv.best_params_)
print("Nilai Rata-Rata (Score) Terbaik: ",LR_cv.best_score_)
print("Akan dipilih Parameter (C) yaitu {0}".format(LR_cv.best_params_))
print("")
LR = LogisticRegression(C = 0.1).fit(X_train, Y_train)
PredictLR = LR.predict(X_test)
print(color.DARKCYAN + "Classification Report dari Metode Logistic Regression :"+color.END)
print(classification_report(Y_test, PredictLR))
print("Test Score:  ", LR.score(X_test,Y_test))
print("Train Score : ", LR.score(X_train,Y_train))


#Metode Decision Tree
print("--------------------------------------------------------------")
print(color.BOLD + color.RED + "Metode Decision Tree"+color.END)
DTC = DecisionTreeClassifier(criterion = 'entropy',random_state=4).fit(X_train,Y_train)
PredictDTC = DTC.predict(X_test)
print(color.DARKCYAN +"Classification Report dari Metode Decision Tree :"+color.END)
print(classification_report(Y_test, PredictDTC))
print("Test Score : ", DTC.score(X_test,Y_test))
print("Train Score : ", DTC.score(X_train,Y_train))
print("")


#Metode Support Vector Machine (SVM)
print("--------------------------------------------------------------")
print(color.BOLD + color.RED + "Metode Support Vector Machine"+color.END)
SVM = svm.SVC(kernel='rbf').fit(X_train,Y_train)
PredictSVM = SVM.predict(X_test)
print(color.DARKCYAN +"Classification Report dari Metode SVM :"+color.END)
print(classification_report(Y_test, PredictSVM))
print("Test Score : ", SVM.score(X_test,Y_test))
print("Train Score : ", SVM.score(X_train,Y_train))
print("--------------------------------------------------------------")

print(color.BOLD + color.BLUE +"Kesimpulan"+color.END)
print("Dengan menganalisa test dan train score yang telah diperoleh tiap model, didapat")
print("1. Model Logistic Regression memiliki test score paling tinggi namun terlihat perbedaan yang cukup jauh dengan train score sehingga memungkinkan terjadinya overfitting")
print("2. Model Decision Tree menghasilkan train score sebesar 1 sehingga memungkinkan terjadinya overfitting")
print("3. Model SVM memiliki test score yang cukup tinggi dan jika dibandingkan dengan train score tidak terdapat perbedaan yang terlalu jauh sehingga kemungkinan terjadi overfitting rendah")
print("Sehingga, model terbaik untuk memprediksi dataset ini adalah "+color.BOLD+"Metode Support Vector Machine"+color.END+".")
```
